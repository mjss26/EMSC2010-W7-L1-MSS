### Help a lost penguin find its way home

* During a winter field trip to Antartica you find a lost penguin.
* It's dark and there is a snow storm!
* You cannot see enough to identify what type of penguin it is.
* But you can measure its height, which is 71 cm
* Should you point in the penguin in the direction of the Emperor or Adélie colony?

What you know:
Heights of the Emperor penguin population are normally distributed:
* Mean = 115 cm
* Standard deviation = 5 cm

Heights of the Adélie penguin population are normally distributed:
* Mean = 60 cm
* Standard deviation = 8 cm

You also have Bayes theorem to help!

$p(A|B) = \frac{p(B|A)⋅p(A)}{p(B)}$,

where:

* $p(A)$ is the prior probability that the penguin is species $A$.
* $p(B|A)$ is the likelihood that a species $A$ penguin has a height of 71 cm.
* $p(B)$ is the probability of observing a 71 cm tall penguin (we won't worry about this, I'll explain why later).
* $p(A|B)$ is the posterior probability that the penguin is species $A$ given that its height is $B$ (this is what we are interested in).


First we'll import the packages we need:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm #Scipy funtion for working with normal distributions
```

```python
prior_E = 0.95 #prior that Emperor penguin is 50/50 (first ex), 95/5 (second ex)
prior_A = 1 - prior_E #so that prior adelie adds up to 1

# can be used to represent the knowledge we already have ie population of penguin colonies, or closeness of each colony etc
```

```python
height = np.linspace(40,140,1001) #1001 equally spaced values between 40 and 140
likelihood_E = norm.pdf(height, loc=115, scale=5) #loc is the mean(115), scale is standard dev (5)
plt.plot(height, likelihood_E, label = 'Emperor')

likelihood_A = norm.pdf(height, loc=60, scale=8) #loc is the mean(60), scale is standard dev (8)
plt.plot(height, likelihood_A, label = 'Adelie')

plt.title('Normal distribution of Emperor and Adalie penguins')
plt.xlabel('Height [cm]')
plt.ylabel('PDF')
plt.legend()

#PDF = probaility distribution function
```

```python
height_penguin = 93 #example height of penguin spotted (71 first ex, 93 second ex)
likelihood_E = norm.pdf(height_penguin, loc=115, scale=5) #loc is the mean(115), scale is standard dev (5)
likelihood_A = norm.pdf(height_penguin, loc=60, scale=8) #loc is the mean(60), scale is standard dev (8)
print(likelihood_E)
print(likelihood_A)
```

```python
#bayes theorem - what is the penguin? Probably?
prob_E = likelihood_E * prior_E
prob_A = likelihood_A * prior_A

prob_total = prob_E + prob_A #we need probabilities to add up to 1

prob_E = prob_E / prob_total
prob_A = prob_A / prob_total

print(prob_E)
print(prob_A)

#posterior probability (must add to 1) - how do we interpret? Probably an Adelie penguin
#note sig figures can be super close to 0
```
